```python
df.columns=["age","blood_pressure","specific_gravity","albumin","sugar","red_blood_cells","pus_cell","pus_cell_clumps","bacteria","blood_glucose_level",
            "blood_urea","serum_creatinine","sodium","potassium","haemoglobin","packed_cell_volume","white_blood_cell_count",
            "red_blood_cell_count","hypertension","diabetes_mellitus","coronary_artery_disease","appetite","peda_edema","anemia","class"]
```

```python
df.columns
```

```python
df.packed_cell_volume.unique()
```

```python
df["packed_cell_volume"]=pd.to_numeric(df["packed_cell_volume"],errors="coerce")
```

```python
df["white_blood_cell_count"]=pd.to_numeric(df["white_blood_cell_count"],errors="coerce")
```

```python
df["red_blood_cell_count"]=pd.to_numeric(df["red_blood_cell_count"],errors="coerce")
```

```python
df.dtypes
```

```python
cat_cols=[col for col in df.columns if df[col].dtype=="object"]
```

```python
cat_cols
```

```python
num_cols=[col for col in df.columns if df[col].dtype!="object"]
```

```python
num_cols
```

```python
# Getting all unique values of cat cols
for col in cat_cols:
  print(f"{col} has {df[col].unique()} values \n")
```

```python
df["diabetes_mellitus"].replace(to_replace={" yes":"yes",'\tno':"no",'\tyes':"yes"},inplace=True)
```

```python
df["diabetes_mellitus"].unique()
```

```python
df["coronary_artery_disease"].replace(to_replace={'\tno':"no"},inplace=True)
df["class"].replace(to_replace={'ckd\t':"ckd"},inplace=True)
```

```python
for col in cat_cols:
  print(f"{col} has {df[col].unique()} values \n")
```

```python
df["class"]=df["class"].map({'ckd':1 , 'notckd':0})
```

```python
med=df[num_cols].median()
df[num_cols]=df[num_cols].fillna(med)
```

```python
mod = df[cat_cols].mode().iloc[0]  # Select the first mode for each column
df[cat_cols] = df[cat_cols].fillna(mod)
```

```python
# Converting categorical columns to numeric using DATA ENCODING
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
for col in cat_cols:
  df[col]=label_encoder.fit_transform(df[col])
```
